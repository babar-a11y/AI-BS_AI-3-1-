```python
#1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 2. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 3. Load the Dataset
file_path = '/content/drive/My Drive/spam (1).csv'

try:
    df = pd.read_csv(file_path, encoding='latin-1')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print(f"Error: File not found at {file_path}. Please check the path.")

# 4. Dataset Description

# 4.1 Display the first 5 rows
print("\n--- First 5 Rows ---")
display(df.head())

# 4.2 Dataset Information (Columns, Data Types, Non-null counts)
print("\n--- Dataset Info ---")
print(df.info())

# 4.3 Check for Missing Values
print("\n--- Missing Values ---")
print(df.isnull().sum())

# 4.4 Descriptive Statistics (Unique values, top occurring, frequency)
print("\n--- Descriptive Statistics ---")
print(df.describe())

# 5. Data Visualization

# 5.1 Count Plot for Category Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Category', data=df, palette='viridis')
plt.title('Distribution of Spam vs Ham Messages')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# 5.2 Print exact counts
print("\n--- Category Counts ---")
print(df['Category'].value_counts())
```
